```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/OPEN RECRUITMENT STAFF BPH HMSD ADYATAMA ITERA 2026 (Jawaban) - Form Responses 1.csv")

df.head()
```

```python
# Konversi timestamp dan hapus duplikat NIM (ambil terakhir)
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df = df.sort_values('Timestamp')
df = df.drop_duplicates(subset='NIM', keep='last')

display(df)
```

```python
cols_keep = [
    'Nama Lengkap',
    'NIM',
    'PILIHAN 1',
    'PILIHAN 2',
    'JIKA IYA MAKA DIVISI MANA YANG ANDA PILIH?'
]

# Ambil semua kolom DIVISI
divisi_cols = [c for c in df.columns if c.startswith('DIVISI')]

# Fungsi ambil nilai divisi yang terisi
def ambil_divisi(row, cols):
    for c in cols:
        if pd.notna(row[c]) and str(row[c]).strip() != '':
            return row[c]
    return None

# Bangun dataframe baru
df_clean = df[cols_keep].copy()

# Divisi pilihan 1
df_clean['Divisi_Pilihan_1'] = df.apply(
    lambda row: ambil_divisi(row, divisi_cols),
    axis=1
)

# Asumsi pilihan 2 jika ada (ambil divisi terisi ke-2)
def ambil_divisi_ke2(row, cols):
    hasil = []
    for c in cols:
        if pd.notna(row[c]) and str(row[c]).strip() != '':
            hasil.append(row[c])
    return hasil[1] if len(hasil) > 1 else None

df_clean['Divisi_Pilihan_2'] = df.apply(
    lambda row: ambil_divisi_ke2(row, divisi_cols),
    axis=1
)

# Rename kolom agar Tableau-friendly
df_clean = df_clean.rename(columns={
    'PILIHAN 1': 'Departemen_Pilihan_1',
    'JIKA IYA MAKA DIVISI MANA YANG ANDA PILIH?': 'Divisi_Alternatif'
})
```
